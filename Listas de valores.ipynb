## Lista

Pode colocar qualquer conteudo nela

```python
lista = [5, 'Rodrigo', 8.9]
```

```python
type(lista)
```

```python
lista.append("Augusto") # Adiciona um elemento no final da lista
```

```python
print(lista)
print(lista[1])
```

## Tupla

```python
tupla = (5, 'Rodrigo', 8.9)
```

```python
type(tupla)
```

```python
tupla.append("Augusto") # Tuplas não podem ser alteradas
```

```python
print(tupla)
```

## Dicionário

```python
dicionario = {'nome': 'Rodrigo', 'cidade': 'São José dos Campos'}
```

```python
type(dicionario)
```

```python
print(dicionario)
```

```python
print(dicionario["nome"])
```

```python
dicionario['ultimo nome'] = 'Rebouças'
```

```python
print(dicionario)
```
